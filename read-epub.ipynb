## Images

```python
list_of_images = []
for image in book.get_items_of_type(ebooklib.ITEM_IMAGE):
    list_of_images.append(image)
    print(image)
```

```python
def generate_image_file(img_name_with_extension, img_content):
    with open(img_name_with_extension, "wb") as file:
        file.write(img_content)
```

```python
for image in list_of_images:
    generate_image_file(image.get_name(), image.get_content())
```

```python
img_index_in_list_of_images = 0

img_name = list_of_images[img_index_in_list_of_images].get_name()
img_type = list_of_images[img_index_in_list_of_images].get_type()
img_content = list_of_images[img_index_in_list_of_images].get_content()

print('Image info:')
print('\tname: ' + str(img_name))
print('\ttype: ' + str(img_type))
print('\tcontent: \n\t\t' + str(img_content))
```

## Stylesheets

### Listing stylesheets in epub

```python
stylesheets = []
for stylesheet in book.get_items_of_type(ebooklib.ITEM_STYLE):
    stylesheets.append(stylesheet)
    print(stylesheet)
```

### Stylesheets content

```python
stylesheet_01 = stylesheets[0].get_content()
stylesheet_02 = stylesheets[1].get_content()
```

```python
print(CSSBeautifier.beautify(stylesheet_01))
print(80*'-')
print(CSSBeautifier.beautify(stylesheet_02))
```

## Creating CSS file

```python
with open("page_css.css", "w", encoding='utf-8') as file:
    file.write(str(CSSBeautifier.beautify(stylesheet_01)))
    
with open("css.css", "w", encoding='utf-8') as file:
    file.write(str(CSSBeautifier.beautify(stylesheet_02)))
```
